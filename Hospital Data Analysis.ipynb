```python
i_u = pd.merge(inquiries,users,on = 'user_id',how = 'inner')
i_u.rename(columns = {'created_at_x':'inquiry_date','country':'user_country','created_at_y':'user_date'}, inplace = True)
i_u_c = pd.merge(i_u,clinics,on = 'clinic_id',how = 'inner')
i_u_c.rename(columns = {'created_at':'clinic_date','updated_at':'clinic_update_date','country':'clinic_country','status':'clinic_status'}, inplace = True)
i_u_c_p = pd.merge(i_u_c,procedures,on = 'procedure_id',how = 'inner')
i_u_c_p.rename(columns = {'created_at':'procedure_date','updated_at':'procedure_update_date'}, inplace = True)
i_u_c_p_b = pd.merge(i_u_c_p,bookings,on = 'inquiry_id',how = 'inner')
i_u_c_p_b.rename(columns = {'created_at':'bookings_date'}, inplace = True)
print(i_u_c_p_b.columns)
i_u_c_p_b
```

```python
df = i_u_c_p_b
df['procedure_cost'] = df['procedure_name'].replace({'Hip Replacement': 10000,'Veneers': 300,'Teeth Whitening': 250,'Breast Implants': 5000 ,'Cancer Treatment': 9000, 'Laser Eye Surgery': 1000 })
df['inquiry_cost'] = df['procedure_name'].replace({'Hip Replacement': 250,'Veneers': 45,'Teeth Whitening': 50,'Breast Implants': 100 ,'Cancer Treatment': 300, 'Laser Eye Surgery': 75 })
#df['abc_commision'] = df['status_y'].replace({'completed':((df['procedure_cost']*30)/100), 'scheduled': 0, 'cancelled': 0})
df['inquiry1'] = df['status'].replace({'completed': 1, 'scheduled': 1, 'cancelled': 0})
df['procedure1'] = df['status'].replace({'completed': 1, 'scheduled': 0, 'cancelled': 0})
df
```
